```python
df_nursing = pd.concat([df_nursing_asir,df_nursing_bahah,df_nursing_hail,df_nursing_jazan,df_nursing_jouf,df_nursing_madinah,df_nursing_makah,df_nursing_najran,df_nursing_north,df_nursing_qassim,df_nursing_riyadh,df_nursing_tabouk,df_nursing_east])
```

```python
df_nursing = df_nursing.reset_index(drop=True)
```

```python
df_nursing
```

```python
df_nursing[df_nursing['المنطقة']=='الباحة']
```

```python
def reshape_arabic(text):
    from arabic_reshaper import reshape
    from bidi.algorithm import get_display
    return get_display(reshape(text))
```

```python
df_nursing['year'] = df_nursing['السنة']
df_nursing['number'] = df_nursing['العدد']
df_nursing['region'] = df_nursing['المنطقة'].apply(reshape_arabic)

order = (
    df_nursing.groupby('region')['number']
    .mean()
    .sort_values(ascending=False)
    .index
)

plt.figure(figsize=(15,8))

sns.lineplot(
    data=df_nursing,
    x='year',
    y='number',
    hue='region',
    style='region',
    hue_order=order,  
    dashes=False,
    markers=True
)

plt.title (reshape_arabic('عدد موظفي التمريض حسب المنطقة والسنوات'))
plt.grid( linewidth=0.5, linestyle='--')
plt.legend(title='region')
plt.savefig('عدد موظفي التمريض حسب المنطقة والسنوات.png',dpi=300)
plt.show()
```

يوضح هذا المخطط الخطي اتجاه عدد موظفي التمريض في مناطق المملكة العربية السعودية من عام 2011 إلى 2022.

**أبرز الملاحظات:**
- تتصدر **الرياض** دائمًا بأعلى عدد من موظفي التمريض. ولو تلاحظ كان هناك زيادة واضحة في أعداد التمريض خلال فترة جائحة كورونا (بلغت الذروة تقريبًا بين 2019-2020) نتيجة الطلب الكبير على خدمات الرعاية الصحية، ثم تبعها انخفاض ملحوظ يُحتمل أنه مرتبط بحالات **الاحتراق الوظيفي** وخروج بعض الكوادر بعد الجائحة.

- بينما بقية المناطق حافظت على أعداد أقل، مع زيادات تدريجية بمرور السنوات.
